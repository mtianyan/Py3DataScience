# 1. axes_style and set_style

```python
%matplotlib inline
```

```python
x = np.linspace(0,14,100)
```

```python
y1 = np.sin(x)
y2 = np.sin(x+2)*1.25
```

```python
def sinplot():
    plt.plot(x, y1)
    plt.plot(x, y2)
```

```python
sinplot()
```

```python
import seaborn as sns
```

```python
sinplot()
```

```python
style = ['darkgrid', 'dark', 'white','whitegrid', 'ticks']
```

#### set_style

```python
sns.set_style(style[3], {'grid.color': 'red'})
```

```python
sinplot()
```

```python
sns.axes_style()
```

```python
sinplot()
```

```python
sns.set()
```

```python
sinplot()
```

```python
sns.set_style('white')
sinplot()
```

```python
sns.set()
```

```python
sinplot()
```
